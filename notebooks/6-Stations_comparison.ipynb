# 1) Data importation, preparation and annual budgets

## 1) 1) Data of Lautaret

```python
# Import data from the ICOS portal 
flux_all_CLt = pd.read_csv('/mnt/data-summer-shared/ICOS_Data_AliseROBERT/ICOSETC_FR-CLt_FLUXNET_HH_L2.csv', na_values=[-9999.0])
```

```python
# Format time data 
flux_all_CLt['TIMESTAMP_START'] = pd.to_datetime(flux_all_CLt['TIMESTAMP_START'], format="%Y%m%d%H%M")
```

```python
# Select only interesting columns in the fluxnet dataset
useful_columns = ['TIMESTAMP_START','NEE_VUT_REF', 'NEE_VUT_REF_QC', 'RECO_NT_VUT_REF', 'GPP_NT_VUT_REF', 'TA_F', 'SWC_F_MDS_1', 'SWC_F_MDS_2', 'VPD_F', 'WS_F', 'LE_F_MDS', 'LE_F_MDS_QC', 'P_F', 'SW_IN_F']
flux_all_CLt = flux_all_CLt[useful_columns] 
```

```python
# Keep data from 2020 to 2024
flux_all_CLt_20_24 = flux_all_CLt[(flux_all_CLt['TIMESTAMP_START'] >= '2020-01-01') & (flux_all_CLt['TIMESTAMP_START'] <= '2024-12-31')]
```

```python
# Convert the carbon flux to gC/m2/30 min
flux_all_CLt_20_24['NEE_VUT_REF'] = flux_all_CLt_20_24['NEE_VUT_REF'] * 12 / 1e6 * 1800
flux_all_CLt_20_24['RECO_NT_VUT_REF'] = flux_all_CLt_20_24['RECO_NT_VUT_REF'] * 12 / 1e6 * 1800
flux_all_CLt_20_24['GPP_NT_VUT_REF'] = flux_all_CLt_20_24['GPP_NT_VUT_REF'] * 12 / 1e6 * 1800
```

```python
# Exclude values of poor quality/nondimensional (QC = 3)
flux_all_CLt_20_24 = flux_all_CLt_20_24[flux_all_CLt_20_24['NEE_VUT_REF_QC'] <= 2]
```

```python
# Import meteorological variables and add it to the flux dataframe
flux_all_Clt_meteo = pd.read_csv('/mnt/data-summer-shared/ICOS_Data_AliseROBERT/ICOSETC_FR-CLt_METEOSENS_L2.csv', na_values=[-9999.0])
flux_all_Clt_meteo['TIMESTAMP_START'] = pd.to_datetime(flux_all_Clt_meteo['TIMESTAMP_START'], format="%Y%m%d%H%M")
flux_all_CLt_20_24 = flux_all_CLt_20_24.merge(flux_all_Clt_meteo[['D_SNOW_1_1_1', 'TA_1_1_1', 'TIMESTAMP_START']], on='TIMESTAMP_START')
```

```python
# Add time columns (month, year, day of year)
flux_all_CLt_20_24['month'] = flux_all_CLt_20_24['TIMESTAMP_START'].dt.month
flux_all_CLt_20_24['year'] = flux_all_CLt_20_24['TIMESTAMP_START'].dt.year
flux_all_CLt_20_24['doy'] = flux_all_CLt_20_24['TIMESTAMP_START'].dt.dayofyear
```

```python
# For snow depth values, replace NaN by 0
flux_all_CLt_20_24['D_SNOW_1_1_1'].replace(np.nan, 0., inplace=True)
```

```python
# Depending on the variables, calculate the daily means (snow depth, temperature) and sums (NEE, Reco, GPP) for each year
sum_CLt_20_24 = (flux_all_CLt_20_24.groupby(['year', 'doy', 'month']).agg({'NEE_VUT_REF' :'sum', 'RECO_NT_VUT_REF':'sum', 'GPP_NT_VUT_REF':'sum', 'D_SNOW_1_1_1':'mean', 'TA_1_1_1': 'mean'}).reset_index())
```

```python
# Determine the annual carbon budgets
budget_CLt_20_24 = flux_all_CLt_20_24.groupby('year')['NEE_VUT_REF'].sum().reset_index()
budget_CLt_20_24
```

```python
estime_CLt = sum_CLt_20_24.groupby('month').mean()
estime_CLt.index = ['Jan', 'Fev', 'Mar', 'Avr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
err_CLt = sum_CLt_20_24.groupby('month').std()
err_CLt.index = ['Jan', 'Fev', 'Mar', 'Avr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

fig,ax = plt.subplots(3,1, figsize = (8,10))
ax[0].bar(estime_CLt.index.values, estime_CLt['TA_1_1_1'], color = 'black', alpha=0.3)
ax[0].errorbar(estime_CLt.index.values, estime_CLt['TA_1_1_1'], yerr=err_CLt['TA_1_1_1'], fmt='none', capsize=6, color = 'black', alpha=0.5)
ax[0].set_ylabel("Temperature (°C)")

ax[1].bar(estime_CLt.index.values, estime_CLt['NEE_VUT_REF'], color = 'red', alpha=0.5)
ax[1].errorbar(estime_CLt.index.values, estime_CLt['NEE_VUT_REF'], yerr=err_CLt['NEE_VUT_REF'], fmt='none', capsize=6, color = 'darkred', alpha=0.5)
ax[1].set_ylabel("CO2 (gC m-2 par jour)")

values = estime_CLt['D_SNOW_1_1_1'].values
errors = err_CLt['D_SNOW_1_1_1'].values

err_inf = np.minimum(errors, values)
err_sup = errors
yerr_asym = [err_inf, err_sup]

ax[2].bar(estime_CLt.index.values, values, alpha=0.5)

# Mask to avoid 0 values
mask = values > 0.01
ax[2].errorbar(estime_CLt.index.values[mask], values[mask],
               yerr=[err_inf[mask], err_sup[mask]],
               fmt='none', capsize=6, alpha=0.5)

ax[2].set_ylabel("Snow depth (m)")
ax[2].set_xlabel("Month")

plt.title('Average annual cycle of temperature (°C), snow depth (m) and CO2 flux (gC/m2/d) at the Col du Lautaret station (period 2020-2024)', y=3.5)
plt.savefig('/home/jovyan//ICOS-M2-internship-alise/data_plots/seasonal_NEE.png', bbox_inches='tight')
plt.show()
```

Monthly averages are represented by bar charts, while standard deviations (daily variability for each month) are symbolized by error bars.
From November to March, average temperatures are negative or close to 0. During this period, the snow cover can exceed one meter: as there is no vegetation and mineralization is slowed but not stopped, the grassland acts as a carbon source (positive average daily sum). Then in April, the snow melts until it disappears entirely, allowing vegetation to grow. The ecosystem will strongly fix carbon during the months of June and July, at the peak of biomass. The summer season remains cool, with average temperatures between 11 and 14°C, but the CO2 flux becomes positive again from August. During the autumn period, the first occasional snowfalls can occur, but until November the snow cover is not necessarily persistent. **The average annual cycle of the Lautaret site is typical of the subarctic climate, with very distinct summer and winter seasons.**

```python
# Replace values outside the range [mini, maxi] with NaN.
def ranger(x,mini,maxi):
    if (x<mini) or (x>maxi):
        x=np.nan
    return x
```

```python
# Mask (set to NaN) all rows in a DataFrame where the solar zenith angle (SZA) is greater than a given threshold.
def sza_cutter(df,sza_lim=85,tz='utc'):
    from sun_pos import sza
    dfc=df.copy()
    SZA = (Series(index=df.index,
                  data=df.index.values)
          ).apply(lambda x:sza(x,tz=tz))
    dfc[SZA>sza_lim]=np.nan
    return dfc
```

```python
# Complete function to clean the NDVI
def clean_ndvi(ndvis,sza_lim=75,plot=False,span=None):
    for col in ndvis.columns:
        ndvis[col] = pd.to_numeric(ndvis[col], errors="coerce")
        ndvis[col]=ndvis[col].apply(lambda x: ranger(x,-1.,1.))
    
    ndvis_clean = sza_cutter(ndvis,sza_lim,'utc')
    
    ndvis_clean.loc['2022-08-26':'2022-08-28']=np.nan
    ndvis_clean.loc['2023-10-01 10:30':'2023-10-04 12',
                   ['NDVI_skye_ref']]=np.nan

    
    if plot:
        compare_plot(ndvis_clean,ndvis,span)
        
    return ndvis_clean
```

At the Lautaret station, annual carbon budgets are positive from 2021 to 2024, reaching 199 g C/m2 per year, which is consistent with data from other studies. The inter-site comparison reveals a less important carbon sink function for the Lautaret ecosystem, that could be attributable to a shorter biomass production period or increased respiration, caused by environmental or ecological factors. Thus, several hypotheses can be formulated to explain the carbon fluxes observed at Lautaret : plant diversity and/ or geomorphology, linked to the site's water characteristics. Differences in soil water content, precipitation and evapotranspiration should be further investigated, for example by spatializing the data in relation to soil properties. Vegetation dynamics (onset of vegetation, senescence, etc.) could respond differently between the 3 sites. Overall, this work highlights the importance of implementing reproducible and universally accessible data and tools, within the context of an international scientific network aiming for exhaustivity and standardization of protocols.
